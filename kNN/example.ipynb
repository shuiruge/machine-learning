### Read and Adjust Raw Data

```python
data_file = open('./datingTestSet.txt', 'r')

raw_training_data = []
for line in data_file:
    raw = line.strip().split('\t')
    rearranged = [[eval(_) for _ in [raw[0], raw[1], raw[2]]], raw[3]]
    raw_training_data.append(rearranged)
    
data_file.close()
```

```python
update_features_boundaries(raw_training_data)

print(features_boundaries)
```

```python
training_data = normalize_raw_training_data(raw_training_data)
```

### Visualizations

```python
conceptions = set([target_concept(_) for _ in training_data])
conceptions
```

```python
def plot(two_dim_points, plot_parameters):
    return plt.plot([_[0] for _ in two_dim_points], [_[1] for _ in two_dim_points], plot_parameters)

def plot_by_feature(two_feature_dimensions):
    """ [Int * Int] -> None
    
    `'didntLike'` is blue point;
    `'smallDoses'` is green point;
    `'largeDoses'` is red point.
    """
    x = two_feature_dimensions[0]
    y = two_feature_dimensions[1]
    plot([[instance(_)[x], instance(_)[y]] for _ in training_data if target_concept(_) == 'didntLike'], 'bo')
    plot([[instance(_)[x], instance(_)[y]] for _ in training_data if target_concept(_) == 'smallDoses'], 'go')
    plot([[instance(_)[x], instance(_)[y]] for _ in training_data if target_concept(_) == 'largeDoses'], 'ro')
    plt.xlabel(str(x))
    plt.ylabel(str(y))
    return None
```

```python
plot_by_feature([0,1])
plt.show()

plot_by_feature([1,2])
plt.show()

plot_by_feature([2,0])
plt.show()
```
